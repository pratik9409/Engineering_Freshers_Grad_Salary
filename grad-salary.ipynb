# Working on specialization column

```python
df.Specialization.unique()
```

```python
df.Specialization.value_counts()
```

```python
# will add specializations whose value_counts are less than 10 in the other column

specialization = df.Specialization.value_counts(ascending=False)
specialization
```

```python
specializationlessthan10 = specialization[specialization<=10]
specializationlessthan10
```

```python
def removespecialessthan10(value):
    if value in specializationlessthan10:
        return 'other'
    else:
        return value
df.Specialization = df.Specialization.apply(removespecialessthan10)
```

```python
# Replacing the -1 values to null
df = df.replace(-1, np.nan)
```

```python
# filling Null values with mean
cols_nan = [column for column in df.columns if df.isna().sum()[column]>0]
for column in cols_nan:
    df[column] = df[column].fillna(df[column].mean())
```


# Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df.Gender = le.fit_transform(df.Gender)
df.Degree = le.fit_transform(df.Degree)
df.Specialization = le.fit_transform(df.Specialization)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor
```

```python
x = df.drop(['Salary'], axis=1)
```

```python
y= df['Salary']
```

```python
sc = StandardScaler()
```

```python
x=sc.fit_transform(x)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, shuffle=True, random_state=0)
```

```python
xgb = XGBRegressor()
```

```python
xgb.fit(X_train, y_train)
```

```python
predictions = xgb.predict(X_test)
```

```python
diff = y_test-predictions
```

```python
plt.figure(figsize=(10,7))
sns.distplot(diff)
```

```python
xgb_r2 = xgb.score(X_test, y_test)
```

```python
print("XGBoost R2 Score", xgb_r2)
```
